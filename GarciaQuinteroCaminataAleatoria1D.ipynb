  Comitas Aleatorias
  

```python
import random
import matplotlib.pyplot as plt

p = 0.5
x = 0
trayectoria = [0]
for i in range(100):
  r = random.random()
  if r < p :
    x = x + 1
  else :
    x = x - 1
  trayectoria.append(x)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
plt.plot(trayectoria)
plt.title("Caminata Aleatoria 1D (100 pasos)")
plt.xlabel("Paso")
plt.ylabel("Posición")
plt.grid(True)
plt.show()
```

Probabilidad MENOR A 0.5

```python
import random
import matplotlib.pyplot as plt

p = 0.2
x = 0
trayectoria = [0]
for i in range(100):
  r = random.random()
  if r < p :
    x = x + 1
  else :
    x = x - 1
  trayectoria.append(x)
plt.figure(figsize=(10,5))
plt.plot(trayectoria)
plt.title("Caminata Aleatoria 1D (100 pasos)")
plt.xlabel("Paso")
plt.ylabel("Posición")
plt.grid(True)
plt.show()
```

# CONCLUSIONES
Cuando la probabilidad p = 0.5, la gráfica muestra que la posición sube y baja sin una tendencia clara, moviéndose de forma irregular alrededor del punto inicial.

Cuando se aumenta la probabilidad a un valor mayor que 0.5, se observa que la trayectoria tiende a crecer con el paso del tiempo, es decir, hay más movimientos hacia arriba que hacia abajo.

Cuando la probabilidad es menor que 0.5, la gráfica muestra una tendencia descendente, ya que predominan los movimientos hacia abajo.
