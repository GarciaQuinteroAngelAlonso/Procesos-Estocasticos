Lo primero que necesitamos es importar las librerias necesarias para el problema, en este caso *Sympy* & *Networkx*, para usar con mas facilidad ambas librerias, solamente las llamaremos *sp* & *nx* respectivamente


```python
import sympy as sp
import networkx as nx
```



Para empezar, creamos una matriz sencilla, llamaremos a la matriz (A), usaremos la libreria sympy y definimos la matriz.

```python
A = sp.Matrix([[4,5],[6,7]])
print (A)
```

*Columna*

```python
A.col(0)
```

*Renglón*

```python
A.row(0)
```

*Entrada Especifica*

```python
A[0,0]
```

Solo que multiplicar ambas matrices, A & B

**Cómo dibujar un diagrama de transición en la librería networkx**

La librería NetworkX permite representar grafos y diagramas de transición, ya importamos la libreria Networkx, importamos de igual manera la librería Matplotlib.
Se usa para dibujar gráficos, en este caso para mostrar el diagrama de transición.

```python
import matplotlib.pyplot as plt
```

*Crear el grafo*

Aquí creamos el grafo donde se guardarán los estados y transiciones.

DiGraph significa Directed Graph (grafo dirigido).

En un grafo dirigido, las flechas tienen dirección.

```python
G = nx.DiGraph()
```

*Agregar Transiciones*

Esta línea agrega una arista (flecha).

```python
G.add_edge("A","B", weight=0.5)
```

*Segunda Transicion*


```python
G.add_edge("A","C", weight=0.5)
```

*Tercera Transicion*

```python
G.add_edge("B","A", weight=1.0)
```

*Cuarta transición*

```python
G.add_edge("C","C", weight=1.0)
```

*Posicion de Nodos*

```python
pos = nx.spring_layout(G)
```

*Dibujar el grafo*

```python
nx.draw(G, pos, with_labels=True, node_size=2000, node_color="lightblue")
```
